```python
#this function replicates the analysis done in Section 10.14 of the textbook

#it takes in the filename, the name of the label, and the features whose dependence plots should
#be made

def run(filename, target, dependence_feat_plot):

    #read file
    df = pd.read_csv(filename)
    
    labels = df[target]
    features = df.drop(target, axis = 1)

    #split dataset and normalize training and testing sets
    x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)

    x_train = (x_train - x_train.mean())/x_train.std()
    x_test = (x_test - x_test.mean())/x_test.std()

    y_train = (y_train - y_train.mean())/y_train.std()

    y_test = (y_test - y_test.mean())/y_test.std()

    
    test_error = []
    train_error =[]

    #here we will graph training and testing error as a
    #function of the number of estimators used by the model
    #for each #of estimators
    for steps in range(0, 1000, 100):
        
        #train a model using that #of estimators
        xg_reg = xgb.XGBRegressor(objective ="reg:pseudohubererror",  learning_rate = 0.1,
                    max_leaves= 6,  n_estimators = steps) 

        #fit the model
        xg_reg.fit(x_train, y_train)

        #have it make predictions on the testing set
        predictions = xg_reg.predict(x_test)

        #calculate the error between the predictions and actual labels
        #and add to train and test error lists

        test_error.append(mean_absolute_error(predictions, y_test))

        
        predictions = xg_reg.predict(x_train)

        train_error.append(mean_absolute_error(predictions, y_train))
    
    
    #plot the training and testing errors as a function of # of estimators
    plt.plot(range(0, 1000, 100), train_error, label = "training error")

    plt.plot(range(0, 1000, 100), test_error, label = "testing error")

    plt.legend(loc= "upper right")
    plt.title("Training and Test Absolute Error")
    plt.xlabel("Number of Iterations")
    plt.ylabel("Absolute Error")

    plt.show()


    #get the relative importance of each feature determined by the model

    feat_imp = xg_reg.feature_importances_
    feat_name = features.columns
    
    #use a dataframe to sort the features by their relative importance
    data = {"Features": feat_name, "Relative Importance": 100*feat_imp/max(feat_imp)}

    data = pd.DataFrame(data)

    data = data.sort_values(by = "Relative Importance", ascending = False)
    
    #make a bar graph of the relative importance of each feature

    plt.barh(data["Features"], data["Relative Importance"])
    plt.title("Feature Importance")
    plt.xlabel("Relative Importance")
    plt.ylabel("Features")
    plt.show()


   # partial dependence plots
   
   #graph the 2-d partial dependence plots 
   
    PartialDependenceDisplay.from_estimator(xg_reg, x_train, ["Income", "Age", "Rooms", "Population"])
    # PartialDependenceDisplay.from_estimator(xg_reg, features, ["Age", "Rooms", "Population", "Income", ("Age", "Population")])
    
    # 3-d partial dependence plot of house value on median age and average occupancy
    # used the 3d example from sklearn's documentation to help: https://scikit-learn.org/stable/auto_examples/inspection/plot_partial_dependence.html
    fig = plt.figure()
    pd_features = ["Age", "Population"]
    pdp = partial_dependence(xg_reg, x_train, pd_features, kind = "average", grid_resolution=10)
    XX, YY = np.meshgrid(pdp["values"][0], pdp["values"][1])
    Z = pdp.average[0].T
    ax = fig.add_subplot(projection="3d")
    fig.add_axes(ax)
    surf = ax.plot_surface(XX, YY, Z, rstride=1, cstride=1, cmap=plt.cm.BuPu, edgecolor="k")
    ax.set_xlabel(pd_features[0])
    ax.set_ylabel(pd_features[1])
    ax.set_zlabel("Partial dependence")
    ax.view_init(elev=22, azim=150)
    plt.colorbar(surf)
    plt.subplots_adjust(top=0.9)
    plt.show()


run("california_housing.csv", "Housevalue", [2,3,5,7, (2, 5)])
```

```python
###################################### STRETCH GOAL #1 ######################################
#splits the dataset into a training and testing set
#and extracts the labels from both sets
#takes in the filename of the dataset and the name of the labels

def prepare_data(filename, target):

    #read in the dataset
    df = pd.read_csv(filename)
    
    #get the labels
    labels = df[target]

    #remove the labels from the feature set
    features = df.drop(target, axis = 1)

    #have 80% of the dataset be for training and 20% for testing


    x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)

    x_test, val_train, y_test, val_test = train_test_split(x_test, y_test, test_size = 0.5)


    #normalize the dataset
    #training and testing set are normalized separately so as to prevent
    #data leakage

    x_train = (x_train - x_train.mean())/x_train.std()


    x_test = (x_test - x_test.mean())/x_test.std()

    y_train = (y_train - y_train.mean())/y_train.std()

    y_test = (y_test - y_test.mean())/y_test.std()

    val_train = (val_train - val_train.mean())/val_train.std()
    val_test = (val_test - val_test.mean())/val_test.std()

    return x_train.to_numpy(), x_test.to_numpy(), y_train.to_numpy(), y_test.to_numpy(), val_train.to_numpy(), val_test.to_numpy()
```

```python
#node class holds best split point as well as an ID
#which can be used to retrieve the region of points contained
#represented by the node

class node:
  def __init__(self, ID, feat, split, val):
    self.ID = ID
    self.feat = feat
    self.split = split
    self.val = val

#tree class contains a list of nodes
#as well as a hashtable which is used
#to get the region for a given node (by their ID)

class Tree:

  def __init__(self, li, hashtable):

    self.li = li
    self.hashtable = hashtable
  
  #adds an entry to the hashtable
  def hash(self, ID, region):

    self.hashtable[ID] = region
  
  #retrieves region for a given ID
  def retrieve(self, ID):
    return self.hashtable[ID]

```

```python

#trains a tree model on the training set
#takes in the number of nodes the tree should have
#the distance function to be used
#and a boolean that says whether this is a regression
#or classification problem

def train(num_Nodes, xtrain, ytrain, dist_func, regression):

  #since our list will start at 1, we will set the first entry to None
  #also since we're preallocating space for our tree, when we use the evaluation
  #function to find the best fit, it's not true that if index*2 > len(tree)
  #that we have reached a leaf node. So to solve this issue, we add a layer of None
  #objects on the bottom of the tree so that if the child of a node is a None object
  #then that node must be a leaf node

  model = Tree([None]*2*2**(math.ceil(math.log(num_Nodes + 1, 2))), dict())

  #make the root node
  root = node(1, 0, 0, 0)
  model.li[1] = root

  #and since it's the root node it contains all of the points in the 
  #training set

  model.hash(1, np.arange(1, np.shape(xtrain)[0]))

  #if we're doing regression
  if(regression):
    #then the value that we assign to any point that ends up on this node 
    #should just be the average of all of the training points in its region
    model.li[1].val = np.average(ytrain)
  else:
    #else we're doing classification and we instead take the mode
    
    model.li[1].val = (stats.mode(ytrain))
  

  
  #for however many nodes we want to add to the tree

  for index in range(1, num_Nodes):

    #find the best feature and split point of the region of the node
    best_feat, best_split, partition1, partition2, value1, value2 = split(model, index, xtrain, ytrain, dist_func, regression)

    #set the node's split point 
    model.li[index].feat = best_feat
    model.li[index].split = best_split

    #instantiate the child nodes of the node and add their values

    child1 = node(index*2, best_feat, best_split, value1)

    model.li[index * 2] = child1
    child2 = node(index*2 +1, best_feat, best_split, value2)

    model.li[index* 2 + 1] = child2

    #update the hashtable with the new entries

    model.hash(index*2 , partition1)
    model.hash(index*2 + 1, partition2)

    
  
  return model


#this function computes the best possible split point for a node
#in a tree
#takes in the current version of the tree model
#the distance function to be used and a boolean that tells us 
#whether or not to do regression
#returns the best feature, the best split point, and the new regions 
#of the child nodes as well as

def split(model, index, xtrain, ytrain, dist_func, regression):

  
  #retrieve the region of the node from its ID

  region = model.retrieve(model.li[index].ID)

  #here we will store the attributes of the node that best improves the performance
  #of the tree
  best_perf = float("inf")
  best_split = 0
  best_feat = 0
  best_val1 = 0
  best_val2 = 0
  #for each feature

  for feat in range(np.shape(xtrain)[1]):
    
    #take all of the points in the parent region and 
    #get their values for the current feature

    bi = [xtrain[point][feat] for point in region]
    
    #sort the points by their feature values
    df = pd.DataFrame({"points": region, "values": bi})
    df = df.sort_values(by = ["values"]) 

    #now we have the region sorted by the current feature value
    region = df["points"].to_numpy()
    
    #these lists will hold the new region for left and right child nodes respectively
    new_region_even = []
    new_region_odd = np.flip(region).tolist()
    
    #for each point in the region
    for point in region:
      
      #make a node in which the split point is the current point

      model.li[index].feat = feat
      model.li[index].split = xtrain[point][feat]

      #for left nodes, points get added if their value is < the split point
      #whereas for right nodes, points get added if their value is > the split point 
      #so as we move through the sorted region, we add the current point to the new_region_even
      #and remove it from new_region_odd

      new_region_even.append(point)
      new_region_odd.pop()
      val1 = 0
      val2 = 0

      #set the new values for these child nodes
      if(regression):
        #since new_region_even starts as an empty list, it's possible that 
        #we try averaging or taking the mode of an empty list which results in an error
        #so if it's empty, we just set val to 0

        if len(new_region_even) != 0:
          val1 = np.average([xtrain[point][feat] for point in new_region_even])
        if len(new_region_odd) != 0:
          val2 = np.average([xtrain[point][feat] for point in new_region_odd])
      else:
        if(len(new_region_even) != 0):
          val1 = stats.mode([xtrain[point][feat] for point in new_region_even])
        if(len(new_region_odd) != 0):
          val2 = stats.mode([xtrain[point][feat] for point in new_region_odd])

      #create the child nodes and add them to the tree
      candidate1 = node(index*2, feat, xtrain[point][feat], val1)
      
      model.li[index*2] = candidate1

      candidate2 = node(index*2 + 1, feat, xtrain[point][feat], val2)
      model.li[index*2 + 1] = candidate2

      
      
      #add the new entries into the hashtable
      model.hash(index*2, new_region_even)
      model.hash(index*2 + 1, new_region_odd)

      #see how the tree performs with the 2 new nodes
      performance = eval(model, ytrain, xtrain, ytrain, dist_func, regression)
      
      #if it did better than the current best performance
      if performance < best_perf:
        #update the attributes of the best node
        best_perf = performance
        best_split = xtrain[point][feat]
        best_feat = feat
        best_val1 = val1
        best_val2 = val2
      
      #take the child nodes out of the tree
      #and remove the index from the hashtable
      model.li[index*2] = None
      model.li[index*2 + 1] = None
  
  
  #get the values for the new region
  bi = [xtrain[point][best_feat] for point in region]

  #partition the region using the optimal split point
  df = pd.DataFrame({"points": region, "values": bi})

  partition1 = df.loc[df["values"] < best_split]
  partition2 = df.loc[df["values"] >= best_split]
  
  partition1 = partition1["points"].to_numpy()
  partition2 = partition2["points"].to_numpy()
 

  return best_feat, best_split, partition1, partition2, best_val1, best_val2


#evaluates the performance of the model
#takes in the model, the features of the testing set, the labels of the testing set,
#a distance function to calculate loss, and a boolean that tells it if we're doing 
#a regression or classification problem

def predict(model, ytrain, features, labels, regression):
  predictions = []

  #for each point in the feature set
  for points in range(np.shape(labels)[0]):
    index = 1

    #run it through the tree to see at which leaf node
    #it ends up

    #while we haven't reached a leaf node
    while( index < len(model.li) and not (model.li[index] is None)):
      
      #get the region of the current node we're at
      region = model.retrieve(model.li[index].ID)

      #get the split point and feature of that node
      feat = model.li[index].feat

      split = model.li[index].split

      #if the current point has a value < the split point
      if features[points][feat] < split:
        #we move down to the left node
        index = 2*index
      else:
        #else we go to the right node
        index = 2*index + 1

    
    #when we get to a leaf node get its value which is the label
    #we assign to the point

    values = ytrain[region]

    predictions.append(model.li[math.floor(index/2)].val)
   
  return predictions

#here we just calculate the error between the values predicted by 
#the model and the ground truth labels

def eval(model, ytrain, features, labels, dist_func, regression):

  #get the predicted values from the model
  predictions = predict(model, ytrain, features, labels, regression)

  #compute the loss function
  return dist_func(predictions, labels)

#this function prints out the tree using recursion
#takes in the tree, the current index, and the depth of the tree that it's at

def printTree(model, index, level=0):

    #if we haven't reached past the end of the tree
    if index < len(model.li) and not (model.li[index] is None):

        #print out the left branch of the remainder of the tree
        printTree(model, index*2, level + 1)
        
        #print out the parent node
        #the number of spaces is used to indicate the depth that the parent node is at
        #we also display the contents in the parent node 
        print(' ' * 4 * level + '-> ' + "feature " + str(model.li[index].feat) + " at " + str(model.li[index].split) + " " + str(len(model.retrieve(model.li[index].ID))))
        
        #print out the right branch of the remainder of the tree
        printTree(model, index*2 + 1, level + 1)


#this function tunes the model by finding the optimal number of nodes to be used
#in the tree and then computes the testing error of the tuned model
#takes in the loss function to be used and a boolean that tells us if we are doing
#regression or classification

def driver(distFunc, regression):

  #get the dataset splits
  xtrain, xtest, ytrain, ytest, valtrain, valtest = prepare_data("california_housing.csv", "Housevalue")

  #find the best number of nodes in the tree
  best_perf = float("inf")
  best_numNodes = 3

  #for each number of nodes
  for numNodes in range(3, 10, 1):

    #train the a decision tree with that many nodes
    model = train(numNodes, xtrain, ytrain, mean_squared_error, regression)
    #evaluate its performance on the validation set
    perf = eval(model, ytrain, valtrain, valtest, mean_squared_error, regression)

    if perf < best_perf:
      best_numNodes = numNodes
      best_perf = perf
  
  #display the optimal number of nodes
  
  print("The optimal number of nodes appears to be: ", best_numNodes)



  #call on train function to train the model using the mean squared error metric
  #and specify that this is a regression problem


  #evaluate the model's performance on the testing set

  model = train(best_numNodes, xtrain, ytrain, mean_squared_error, regression)
  print("On the testing set, the tuned model achieves an error of: ", eval(model, ytrain, xtest, ytest, mean_squared_error, regression))

  print("\n\n")
  print("And here's a tree diagram of our model: ")
  print("\n\n")
  printTree(model, 1)


driver(mean_squared_error, True)
```
